```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# loading iris dataset using sklearn library
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame

df.head()
```

```python
df["species"] = df["target"].map(dict(enumerate(iris.target_names)))
df = df.drop(columns=["target"])
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
df.info()
```

```python
# displaying descriptive statistics
df.describe(include="all")
```

```python
# scatter plot of sepal length vs sepal width.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="sepal length (cm)", y="sepal width (cm)", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# visualizing histograms for distributions
df.hist(figsize=(10, 8), bins=15)
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
# visualizing box plots to spot outliers and spreads
df_melted = df.melt(id_vars="species", var_name="feature", value_name="value")
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_melted, x="feature", y="value")
plt.xticks(rotation=45)
plt.title("Box Plots of Iris Features")
plt.show()
```
